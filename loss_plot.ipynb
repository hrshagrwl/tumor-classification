```python
import pickle
from matplotlib import pyplot
```

```python
pkl_path = "dipg_vs_normals_training_history.pkl"
```

```python
history = pickle.load(open(pkl_path, "rb"))
```

```python
history
```

```python
pyplot.plot(history['train_loss'])
pyplot.plot(history['val_loss'])
pyplot.title('model train vs validation loss')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(['train', 'validation'], loc='upper right')
pyplot.show()
```

```python
pyplot.plot(history['train_acc'])
pyplot.plot(history['val_acc'])
pyplot.title('model train vs validation loss')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(['train', 'validation'], loc='upper right')
pyplot.show()
```
